# Panorama do COVID-19 no Brasil

Os coronavírus são uma grande família de vírus comuns em muitas espécies diferentes de animais, incluindo camelos, gado, gatos e morcegos. Raramente, os coronavírus que infectam animais podem infectar pessoas, como exemplo do MERS-CoV e SARS-CoV. Recentemente, em dezembro de 2019, houve a transmissão de um novo coronavírus (SARS-CoV-2), o qual foi identificado em Wuhan na China e causou a COVID-19, sendo em seguida disseminada e transmitida pessoa a pessoa.

<center><img alt="Colaboratory logo" width="50%" src="https://image.freepik.com/vetores-gratis/covid-19-poster-com-icone-de-virus_1142-7401.jpg"></center>

A COVID-19 é uma doença causada pelo coronavírus, denominado SARS-CoV-2, que apresenta um espectro clínico variando de infecções assintomáticas a quadros graves. De acordo com a Organização Mundial de Saúde, a maioria (cerca de 80%) dos pacientes com COVID-19 podem ser assintomáticos ou oligossintomáticos (poucos sintomas), e aproximadamente 20% dos casos detectados requer atendimento hospitalar por apresentarem dificuldade respiratória, dos quais aproximadamente 5% podem necessitar de suporte ventilatório.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Por se tratar de uma doença muito nova, os dados sobre o corona vírus, principalmente no Brasil, são muito conflitantes pois podem provir dos governos estaduais, do federal ou de fontes não oficiais.

Nesse projeto irei optar por usar os dados provenientes do repositório do site [Our World in Data](https://ourworldindata.org), que pode ser acessado [neste link](https://github.com/owid/covid-19-data/tree/master/public/data).

### Dicionário de Variáveis

* `iso_code`: Códigos de países de três letras - 3166-1 alfa-3 ISO

* `continent`: Continente da localização geográfica

* `location`: Localização geográfica

* `date`: Data de observação

* `total_cases`: Total de casos confirmados de COVID-19

* `new_cases`: Novos casos de COVID-19 confirmados

* `new_cases_smoothed`: Novos casos de COVID-19 confirmados (análise de 7 dias)

* `total_deaths`: Total de mortes atribuídas a COVID-19

* `new_deaths`: Novas mortes atribuídas a COVID-19

* `new_deaths_smoothed`: Novas mortes atribuídas a COVID-19 (análise de 7 dias)

* `total_cases_per_million`: Total de casos confirmados de COVID-19 por 1.000.000 de pessoas

* `new_cases_per_million`: Novos casos confirmados de COVID-19 por 1.000.000 de pessoas

* `new_cases_smoothed_per_million`: Novos casos confirmados de COVID-19 por 1.000.000 de pessoas (análise de 7 dias)

* `total_deaths_per_million`: Total de mortes atribuídas a COVID-19 por 1.000.000 de pessoas

* `new_deaths_per_million`: Novas mortes atribuídas ao COVID-19 por 1.000.000 de pessoas

* `new_deaths_smoothed_per_million`: Novas mortes atribuídas a COVID-19 por 1.000.000 de pessoas (análise de 7 dias)

* `reproduction_rate`: Estimativa em tempo real da taxa de reprodução efetiva (R) de COVID-19. Veja https://github.com/crondonm/TrackingR/tree/main/Estimates-Database

* `icu_patients`: Número de pacientes de COVID-19 em unidades de cuidados intensivos (UTIs) num dado dia

* `icu_patients_per_million`: Número de pacientes de COVID-19 em unidades de cuidados intensivos (UTIs) num dado dia por 1.000.000 de pessoas

* `hosp_patients`: Número de pacientes de COVID-19 no hospital em um determinado dia

* `hosp_patients_per_million`: Número de pacientes de COVID-19 no hospital em um determinado dia por 1.000.000 de pessoas

* `weekly_icu_admissions`: Número de pacientes de COVID-19 recém-admitidos em unidades de cuidados intensivos (UTIs) em uma dada semana

* `weekly_icu_admissions_per_million`: Número de pacientes de COVID-19 recém-admitidos em unidades de terapia intensiva (UTIs) em uma determinada semana por 1.000.000 de pessoas

* `weekly_hosp_admissions`: Número de pacientes de COVID-19 recém-admitidos em hospitais em uma semana

* `weekly_hosp_admissions_per_million`: Número de pacientes de COVID-19 recém-admitidos em hospitais em uma determinada semana por 1.000.000 de pessoas

* `total_tests`: Total de Testes para COVID-19

* `new_tests`: Os novos testes para COVID-19 (somente calculada para dias consecutivos)

* `total_tests_per_thousand`: Total de Testes para COVID-19 por 1.000 pessoas

* `new_tests_per_thousand`: Novos testes para COVID-19 por 1.000 pessoas

* `new_tests_smoothed`: Os novos testes para COVID-19 (análise de 7 dias). Para os países que não relatam os dados de teste em uma base diária, assumimos que os testes mudaram igualmente em uma base diária sobre quaisquer períodos em que nenhum dado foi reportado. Isso produz uma série completa de figuras diárias, que é então calculada em média ao longo de uma janela móvel de 7 dias

* `new_tests_smoothed_per_thousand`: Os novos testes para COVID-19 (análise de 7 dias) por 1.000 pessoas

* `positive_rate`: Os testes de COVID-19 que são positivos, dados como uma média móvel de 7 dias (este é o inverso da `tests_per_case`)

* `tests_per_case`: Testes realizados por novo caso confirmado de COVID-19, dados como uma média móvel de 7 dias (este é o inverso da `positive_rate`)

* `tests_units`: Unidades usadas pela localização para relatar seus dados de testes

* `total_vaccinations`: Número total de doses de vacinação de COVID-19 administradas

* `people_vaccinated`: Número total de pessoas que receberam pelo menos uma dose da vacina

* `people_fully_vaccinated`: Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação

* `new_vaccinations`: Novas doses de vacinação de COVID-19 administradas (somente calculadas para dias consecutivos)

* `new_vaccinations_smoothed`: Novas doses de vacinação de COVID-19 administradas (análise de 7 dias). Para os países que não relatam os dados de vacinação em uma base diária, assumimos que as vacinações mudaram igualmente em uma base diária sobre quaisquer períodos em que nenhum dado foi reportado. Isso produz uma série completa de figuras diárias, que é então calculada em média ao longo de uma janela móvel de 7 dias

* `total_vaccinations_per_hundred`: Número total de doses de vacinação de COVID-19 administradas por 100 pessoas na população total

* `people_vaccinated_per_hundred`: Número total de pessoas que receberam pelo menos uma dose da vacina por cada 100 pessoas no total da população

* `people_fully_vaccinated_per_hundred`: Número total de pessoas que receberam todas as doses prescritas pelo protocolo de vacinação por 100 pessoas na população total

* `new_vaccinations_smoothed_per_million`: Novas doses de vacinação de COVID-19 administradas (análise de 7 dias) por 1.000.000 de pessoas na população total

* `stringency_index`: Índice de Restrição de Resposta do Governo: medida composta com base em 9 indicadores de resposta incluindo o fechamento de escolas, fechamento de locais de trabalho, e proibição de viajar, escalonados para um valor de 0 a 100 (100 = resposta mais estrita)

* `population`: População em 2020

* `population_density`: Número de pessoas divididas por área de terra, medido em quilômetros quadrados, do ano mais recente disponível

* `median_age`: A idade média da população, projeção das Nações Unidas para 2020

* `aged_65_older`: Parte da população que tem 65 anos ou mais, do ano mais recente disponível

* `aged_70_older`: Parte da população que tem 70 anos ou mais em 2015

* `gdp_per_capita`: Produto interno bruto em paridade de poder aquisitivo (dólares internacionais constantes de 2011), do ano mais recente disponível

* `extreme_poverty`: Parcela da população vivendo em extrema pobreza, do ano mais recente disponível desde 2010

* `cardiovasc_death_rate`: Taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas)

* `diabetes_prevalence`: prevalência de diabetes (% da população com idades entre 20 e 79) em 2017

* `female_smokers`: Parcela de mulheres que fumam, do ano mais recente disponível

* `male_smokers`: Parcela de homens que fumam, do ano mais recente disponível

* `handwashing_facilities`: Parcela da população com instalações básicas para lavagem das mãos no local, do ano mais recente disponível

* `hospital_beds_per_thousand`: camas hospitalares por 1.000 pessoas, do ano mais recente disponível desde 2010

* `life_expectancy`: A expectativa de vida ao nascer em 2019

* `human_development_index`: Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano - uma vida longa e saudável, conhecimento e um padrão de vida decente. Valores para 2019, importados de http://hdr.undp.org/en/indicators/137506



### Importando os dados do COVID-19


```
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Importando o dataset COVID-19 para um DataFrame
df= pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

## Análise Exploratória dos Dados

Agora que importamos os dados do dataset, vamos verificar o que esses dados podem mostrar.

### Análise Exploratória Inicial
Começamos verificando o tamanho do nosso DataFrame e vendo as primeiras entradas.


```
# Checando o tamanho DataFrame
print('Entradas: ',df.shape[0], "\n"
      'Variáveis: ', df.shape[1])

# Checando as 5 primeiras entradas
df.head()
```

```
# Extraindo o nome das colunas
df.columns
```

Na hora de analisar os dados, é importante checar os tipos das variáveis utilizando o método `dtypes`:

```
# Checando os tipos das variáveis
df.dtypes
```

Pudemos perceber que a coluna `date` é uma variável em formato objeto, por isso vamos transformá-la em formato datetime.

```
# Transformando a coluna date para datetime
df.date = pd.to_datetime(df.date)
```

Outro passo importante é verificar os dados ausentes para saber não só quais estão ausentes, mas também o porquê deles estarem ausentes.

```
# Verificando a porcentagem de dados ausentes
round((df.isnull().sum() / df.shape[0]).sort_values(ascending=False),2)
```

No caso desse DataFrame, os dados ausentes podem significar dados que não foram providos pelas respectivas fontes dos países ou dias em que aquelas determinadas variáveis foram nulas.

Vamos ver, para a data mais atual, quais os países que têm mais casos e mortes por COVID-19.

```
# Verificando qual a data mais atual
df.date.max()
```

```
# Verificando os países com mais casos
excluir_continentes= df.copy()
excluir_continentes= excluir_continentes[excluir_continentes['continent'].notna()]

excluir_continentes.loc[df.date == '2021-03-07'].sort_values(by="total_cases", ascending=False)[0:5]
```

Os países que lideram o ranking de casos em **7 de Março de 2021** são:
* 1º: Estados Unidos - 28.998.834 casos
* 2º: Índia - 11.229.398 casos
* 3º: Brasil - 11.019.344 casos

```
# Verificando os países com mais mortes
excluir_continentes.loc[df.date == '2021-03-07'].sort_values(by="total_deaths", ascending=False)[0:5]
```

Os países que lideram o ranking de mortes em **7 de Março de 2021** são:
* 1º: Estados Unidos - 525.031 mortes
* 2º: Brasil - 265.411 mortes
* 3º: México - 190.604 mortes

Para visualizar melhor o número de mortes, vamos plotar um gráfico de barras:

```
# Plotando gráfico de barras
top_5_total_deaths = df.loc[df.date == '2021-03-07', ['location', 'total_deaths']].sort_values(by="total_deaths", ascending=False)[1:11]

fig, ax = plt.subplots(figsize=(15,5))
ax.bar('location', 'total_deaths', data=top_5_total_deaths)

plt.tight_layout()
```

Através do gráfico podemos perceber que o país com maior número de mortes (Estados Unidos) **tem mais mortes do que continentes** como a América do Sul e a Ásia.

Abaixo vamos plotar um gráfico de dispersão considerando `gdp_per_capita` (PIB dos países) por `total_deaths` (número total de mortes), analisando apenas a data mais recente (**7 de Março de 2021**). É importante lembrar que esse dado **não é algo conclusivo**, pois envolve fatores como proporção de testes, etc.

```
# Plotando gráfico de dispersão
df_scatter= df.loc[df.date == '2021-03-07']

fig,ax = plt.subplots(figsize=(10,5))

df_scatter.plot.scatter('gdp_per_capita','total_deaths',ax=ax,s=50)

plt.tight_layout()
```

### Progressão do vírus

Para acompanhar como o COVID-19 está progredindo ao longo do tempo, vamos plotar um gráfico de linhas para mostrar a evolução do número total de casos no mundo e um para mostrar a evolução do número total de mortes.

```
# Plotando gráfico de linhas da evolução do total de casos
df.loc[df.location == "World", "total_cases"].reset_index(drop=True).plot(figsize=(10,8));
```

```
# Plotando gráfico de linhas da evolução do total de mortes
df.loc[df.location == "World", "total_deaths"].reset_index(drop=True).plot(figsize=(10,8));
```

Podemos perceber que o gráfico do total de mortes foi mais **"explosivo"**, enquanto o gráfico do total de casos foi mais **progressivo**.

### Análise Exploratória para o Brasil

Agora vamos fazer uma análise voltada para o Brasil em relação ao Corona Vírus, por isso será feita uma cópia do DataFrame que utilize apenas os dados do país.

```
# Criando uma cópia do DataFrame
df_brasil = df.loc[df.location == "Brazil"].copy()

# Checando as primeiras entradas do novo DataFrame
df_brasil.head()
```

```
# Data máxima
df_brasil.date.max()
```

Acima pudemos verificar que o DataFrame contém dados de **26-02-2020** a **07-03-2021**.

Em seguida, vamos verificar em que data foram registrados o primeiro caso e a primeira morte no Brasil.

```
# Identificando quando foi registrado o primeiro caso
df_brasil.loc[df_brasil['new_cases'] != 0.0].head(1)
```

O primeiro caso de COVID-19 no Brasil foi registrado em **26-02-2020**.

```
# Identificando quando foi registrada a primeira morte
df_brasil.loc[df_brasil['new_deaths'] == 1.0].head(1)
```

A primeira morte por COVID-19 no Brasil foi registrada em **17-03-2020**.

É interessante verificar quanto tempo demorou entre o primeiro caso e a primeira morte.

```
# Verificando tempo entre primeiro caso e primeira morte
from datetime import date
delta = df_brasil.date.loc[9452] - df_brasil.date.loc[9432]
print(delta)
```

Vemos que se passaram **20 dias** entre o primeiro caso registrado e a primeira morte registrada.

Para finalizar, vamos plotar um gráfico para comparar a evolução dos casos e um para comparar a evolução das mortes no Brasil.

```
# Plotando gráficos para comparação
fig, ax = plt.subplots()

df_brasil["total_cases"].plot(kind="line", ax=ax, figsize=(18,6))

ax.set(xlabel="Tempo",
       ylabel="N° de casos",
       title = "Total de Casos no Brasil")
ax.axes.ticklabel_format(axis='y',style='plain')

plt.tight_layout()
```
